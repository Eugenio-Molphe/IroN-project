```python
# Import libraries
import pandas as pd
import os
```

Here I explore the results of the double check made by Nico and then I make a list with the sequences found to be effectively putative iroN on a normal Blast and SwissProt (I decided to only use Blast). And the tested sequences are the ones reported by Prodigal to be the analyzed gene.

```python
#### Here I read the results of the Blast ####

path = '/mnt/mnemo6/nnaepf/for_eugenio/check_annotation/results_NR/result_sequences_ironDB.blast.filter'

df = pd.read_csv(path, delimiter='\t', header=None)

# Modify it by assigning headers and removing useless info in the last column
df.columns = ['query','target','evalue','bits','pident','nident','alnlen','mismatc','qstart','qend','tstart','tend','theader']

def process_string(s):
    leftSideRemoved = s.split(' ')[1:]
    leftSideRemoved = ' '.join(leftSideRemoved)
    rightSideRemoved = leftSideRemoved.split('[')[0]
    return rightSideRemoved
df['theader'] = df['theader'].apply(process_string)

# Now let's sort the values by the guessed protein
df = df.sort_values(by='theader')

# Now let's remove the prodigal results
df = df[df['query'].str.contains('prodigal') == True]

print(df)
```

```python
### Here I check which proteins are found
print(df['theader'].unique())
```

```python
### Now time to filter out the results that doesn't make sense

df1 = df[df['theader'] == 'acyl-CoA esterase ']

'''
The following ones have to be removed:
DJ-1/PfpI family protein 
IS1-like element transposase 
IroE 
MULTISPECIES: acyl-CoA thioester hydrolase YciA
MULTISPECIES: anti-adapter protein IraP  
Salmochelin siderophore protein IroE 
Shikimate 5-dehydrogenase-like protein HI_0607 
Uncharacterised protein 
acyl-CoA esterase 
acyl-CoA thioester hydrolase YciA 
acyl-CoA thioester hydrolase YciA, partial 
alpha/beta hydrolase 
alpha/beta hydrolase, partial 
catecholate siderophore esterase IroE 
enterochelin esterase 
hydrolase, partial 
hypothetical protein 
hypothetical protein PPOKNCAL_00019 
ligand-gated channel protein 
septation protein IspZ 
shikimate 5-dehydrogenase 
'''

filteredDf = df[(df['theader'].str.contains('acyl-CoA') == False) & (df['theader'] != 'DJ-1/PfpI family protein ')
                & (df['theader'] != 'DJ-1/PfpI family protein ') & (df['theader'] != 'IS1-like element transposase ')
                & (df['theader'].str.contains('IroE ') == False) & (df['theader'] != 'MULTISPECIES: anti-adapter protein IraP  ')
                & (df['theader'].str.contains('5-dehydrogenase') == False) & (df['theader'] != 'Uncharacterised protein ')
                & (df['theader'].str.contains('hydrolase') == False) & (df['theader'].str.contains('hypothetical') == False)
                & (df['theader'] != 'ligand-gated channel protein ') & (df['theader'] != 'septation protein IspZ ')]
print(filteredDf)
```

```python
### Now let's create a list of the queries that we want to keep

queriesList = filteredDf['query'].tolist()

# Remove the prodigal header and the things after the point
for idx in range(len(queriesList)):
    queriesList[idx] = queriesList[idx].strip('prodigal_')
    if queriesList[idx][0].isdigit():
        queriesList[idx] = queriesList[idx].split('.')[1]
    else:
        queriesList[idx] = queriesList[idx].split('_')[0]

queriesSet = set(queriesList)

print(queriesList[0:20])
print(len(queriesSet))
```

```python
#### Here I read the results of the Blast ####

path = '/mnt/mnemo6/nnaepf/for_eugenio/escherichia_annot/escherichia_prodigal_results.faa.blast.filter.processed'

df = pd.read_csv(path, delimiter='\t', header=None)

# So let's check the proteins that were found
proteinsFound = df[12].unique()
print(proteinsFound)

# I'm going to accept the following proteins:
'''
'siderophore salmochelin receptor IroN '
'MULTISPECIES: siderophore salmochelin receptor IroN '
'TonB-dependent receptor, partial '
'siderophore salmochelin receptor IroN, partial '
'TonB-dependent siderophore receptor,IroN protein '
'TonB-dependent siderophore receptor '
'IroN '
'tonB-dependent outer membrane siderophore receptor domain protein '
'TonB-dependent receptor plug domain-containing protein, partial '
'''
```

```python
# There's some contig names, I'll search them in the genomes dataset
# So I create a dictionary with the following structure:
# {header_code: NCBI_taxonomic_code}

# Sample data: list of FASTA file names
directory = '/mnt/mnemo5/eugenio/IroN_project/Files/10_Escherichia/Genomes_w_iroN'

allFastaFiles = os.listdir(directory)

# Create an empty dictionary to hold the mappings
headerToNCBIcode = {}

# Iterate through each FASTA file
for file_name in allFastaFiles:
    # Check if file exists to avoid errors
    with open(directory + '/' + file_name, 'r') as file:
        # Let's extract the codes in the headers of the fasta file
        lines = file.readlines()
        headersFastaFile = [line for line in lines if line.startswith('>')]
        headersFastaFile = [header[1:].strip() for header in headersFastaFile]
        headersFastaFile = [header.split(' ')[0] for header in headersFastaFile]
        # Add the mapping to the dictionary
        for header in headersFastaFile:
            headerToNCBIcode[header] = file_name

# Only get the NCBI taxonomic code for the file names
for key in headerToNCBIcode:
    headerToNCBIcode[key] = headerToNCBIcode[key].split('_')[0] + '_' + headerToNCBIcode[key].split('_')[1]

# Print the results
print(headerToNCBIcode)
```

```python
# Time to filter out the annotations that are not for the iroN protein
dfFiltered = df[(df[12] == 'siderophore salmochelin receptor IroN ') | (df[12] == 'MULTISPECIES: siderophore salmochelin receptor IroN ')
                | (df[12] == 'TonB-dependent receptor, partial ') | (df[12] == 'siderophore salmochelin receptor IroN, partial ')
                | (df[12] == 'TonB-dependent siderophore receptor,IroN protein ') | (df[12] == 'TonB-dependent siderophore receptor ')
                | (df[12] == 'IroN ') | (df[12] == 'tonB-dependent outer membrane siderophore receptor domain protein ')
                | (df[12] == 'TonB-dependent receptor plug domain-containing protein, partial ')]

# And obtain the list of the genomes that contain it
genomes = dfFiltered[0].unique().tolist()
print(genomes)

# We change the name of the genomes to the NCBI taxonomic code

for idx in range(len(genomes)):
    if genomes[idx].startswith('GCA_'):
        genomes[idx] = genomes[idx].split('_')[0] + '_' + genomes[idx].split('_')[1]
    else:
        contigName = genomes[idx].split('_')[0]
        genomes[idx] = headerToNCBIcode[contigName]


print(genomes)

# We make the list with the genomes into a set to remove duplicates
genomesSet = set(genomes)
print(genomesSet)

# Now we save the results in a file

pathGenomes = '/mnt/mnemo5/eugenio/IroN_project/Files/10_Escherichia/genomes_w_double_checked_iroN.txt'
with open(pathGenomes, 'w') as f:
    for item in genomesSet:
        f.write("%s\n" % item)
```

```python
# Time to filter out the annotations that are not definitely iroN protein

# Open the file with the genomes that have the iroN protein
with open('/mnt/mnemo5/eugenio/IroN_project/Files/10_Escherichia/genomes_w_double_checked_iroN.txt', 'r') as f:
    genomes = f.readlines()
    for i in range(len(genomes)):
        genomes[i] = genomes[i].strip('\n')

# Open the file with the promoters
promoters = pd.read_csv('/mnt/mnemo5/eugenio/IroN_project/Files/10_Escherichia/promoters_df_with_immunity_genes.csv', sep='\t')

# Filter the promoters DataFrame to obtain only the promoters that are for the iroN gene
promotersFiltered = promoters[promoters['Genome'].isin(genomes)]

print(promotersFiltered)

# Save the results in a file
path = '/mnt/mnemo5/eugenio/IroN_project/Files/10_Escherichia/promoters_df_with_iroN.csv'
promotersFiltered.to_csv(path, sep='\t', index=False)
```
